THE NEXT STEP IS TO ELIMINATE THE COLUMNS WITH OVER 60% OF MISSING VALUES TO UNDERSTAND WELL ON WHICH COLUMNS WE CAN PERFORM A MEANINGFUL ANALYISIS 

```python
#GENDER
missing_percentage = (df_gender.isnull().sum() / len(df_gender)) * 100
print("\nPercentage of missing values per column:\n", missing_percentage)

#Identify columns to drop (those with more than 60% missing values)
columns_to_drop = missing_percentage[(missing_percentage > 60)].index
print("\nColumns to drop (with > 60% missing values):\n", columns_to_drop)

gender_cleaned = df_gender.drop(columns=columns_to_drop)
print(f"\nColumns with more than 60% missing values dropped.")
print(f"Shape of the dataset after dropping columns: {gender_cleaned.shape}")

print("\nFirst 5 rows of the cleaned dataset:")
print(gender_cleaned.head())

print("\nInformation about the cleaned dataset:")
print(gender_cleaned.info())
```

```python
#EDUCATION
missing_percentage = (df_education.isnull().sum() / len(df_education)) * 100
print("\nPercentage of missing values per column:\n", missing_percentage)

columns_to_drop = missing_percentage[(missing_percentage > 60)].index
print("\nColumns to drop (with > 60% missing values):\n", columns_to_drop)

education_cleaned = df_education.drop(columns=columns_to_drop)
print(f"\nColumns with more than 60% missing values dropped.")
print(f"Shape of the dataset after dropping columns: {education_cleaned.shape}")

print("\nFirst 5 rows of the cleaned dataset:")
print(education_cleaned.head())

print("\nInformation about the cleaned dataset:")
print(education_cleaned.info())
```

```python
#POVERTY
missing_percentage = (df_poverty.isnull().sum() / len(df_poverty)) * 100
print("\nPercentage of missing values per column:\n", missing_percentage)

columns_to_drop = missing_percentage[(missing_percentage > 60)].index
print("\nColumns to drop (with > 60% missing values):\n", columns_to_drop)

poverty_cleaned = df_poverty.drop(columns=columns_to_drop)
print(f"\nColumns with more than 60% missing values dropped.")
print(f"Shape of the dataset after dropping columns: {poverty_cleaned.shape}")

print("\nFirst 5 rows of the cleaned dataset:")
print(poverty_cleaned.head())

print("\nInformation about the cleaned dataset:")
print(poverty_cleaned.info())
```

```python
#GENDER
missing_percentage = (df_social_dev.isnull().sum() / len(df_social_dev)) * 100
print("\nPercentage of missing values per column:\n", missing_percentage)

columns_to_drop = missing_percentage[(missing_percentage > 60)].index
print("\nColumns to drop (with > 60% missing values):\n", columns_to_drop)

social_dev_cleaned = df_social_dev.drop(columns=columns_to_drop)
print(f"\nColumns with more than 60% missing values dropped.")
print(f"Shape of the dataset after dropping columns: {social_dev_cleaned.shape}")

print("\nFirst 5 rows of the cleaned dataset:")
print(social_dev_cleaned.head())

print("\nInformation about the cleaned dataset:")
print(social_dev_cleaned.info())
```

WE HAVE OBTAINED THE FINAL DATASET FOR EACH OF THE FOUR TOPICS

Now, we can proceed to merge the datasets based on 'Country Name' and 'Year' and Country Code

```python
#Define the key columns for merging
merge_keys = ['Country Name', 'Country Code', 'Year']

#Merge gender and education
merged_df_1 = pd.merge(gender_cleaned, education_cleaned, on=merge_keys, how='outer')
print(f"Shape after merging gender and education: {merged_df_1.shape}")

#Merge the result with poverty
merged_df_2 = pd.merge(merged_df_1, poverty_cleaned, on=merge_keys, how='outer')
print(f"Shape after merging with poverty: {merged_df_2.shape}")

#Merge the result with social development
final_merged_df = pd.merge(merged_df_2, social_dev_cleaned, on=merge_keys, how='outer')
print(f"Shape after merging with social development: {final_merged_df.shape}")

print("\n--- Final Merged DataFrame ---")
print("Shape:", final_merged_df.shape)
print("\nFirst 5 rows:")
print(final_merged_df.head())
print("\nLast 5 rows:")
print(final_merged_df.tail())
print("\nInfo:")
final_merged_df.info()

output_folder = '/Users/joelsonnino/Desktop/progetto visual/dataset_cleaned'
final_merged_df.to_csv(f"{output_folder}/final_merged_dataset.csv", index=False)
print(f"\\nFinal merged dataset saved to {output_folder}/final_merged_dataset.csv")

final_merged_df
```
